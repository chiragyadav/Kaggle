```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('Indicators.csv')
```

```python
def plot_indicator(indicator_code,ylabel=None,title=None):
    indicator_name= str(df[df.IndicatorCode==indicator_code]['IndicatorName'].unique()[0])
    if(title==None):
        title=indicator_name.split('(')[0]
    if(ylabel==None):
        if(len(indicator_name.split('('))>1):
            ylabel= indicator_name.split('(')[1].replace(')','')
        else:
            ylabel = indicator_name
        
    df_indicator = df[(df.CountryName=='India')&(df.IndicatorCode==indicator_code)]
    plt.plot(df_indicator.Year,df_indicator.Value,'o-',color='g')
    plt.xlabel('Years')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
indicators_list=['SL.TLF.TERT.FE.ZS','SL.TLF.TERT.MA.ZS','SL.UEM.TERT.ZS','SP.URB.TOTL','SP.RUR.TOTL','SP.POP.SCIE.RD.P6']
```

```python
for indicator in indicators_list:
    plot_indicator(indicator)
```
